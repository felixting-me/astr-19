These cells define functions for sin(x) and cos(x). The functions will be used to compute values of sin(x) and cos(x) for x between 0 and 2.

```python
import numpy as np

def sin_function(x):
    return np.sin(x)
```

This function returns the cos(x).

```python
def cos_function(x):
    return np.cos(x)
```

```python
x_values = np.linspace(0, 2, 1000)

sin_values = sin_function(x_values)
cos_values = cos_function(x_values)
```

```python
print("x      sin(x)      cos(x)")
print("-----------------------------")

for i in range(10):
    print(x_values[i], sin_values[i], cos_values[i])
```
