```python
a = 3.5
b = 2.1

x = 10
y = 4

sum_result = a + b
print("sum:", sum_result)
print("data type:", type(sum_result))

diff_result = x - y
print("difference:", diff_result)
print("data type:", type(diff_result))

product_result = a * x
print("product:", product_result)
print("data type:", type(product_result))
```
